```python
from google.colab import files
myfile = files.upload()
```

```python
import io
import pandas as pd
iris = pd.read_csv(io.BytesIO(myfile['caffeine.csv']))
import seaborn as sns

sns.set_theme()
sns.set(rc={"figure.dpi":300}) #이미지 품질 설정
sns.set(rc={"figure.figsize":(6,3)}) ## 크기설정

sns.scatterplot(x="Volume (ml)", y ="Caffeine (mg)", data=iris, hue ='type')
```

```python
## 히스토그램 histogram

sns.histplot( y= "Volume (ml)", data = iris)
```

```python
sns.histplot( x= "Volume (ml)", data = iris)
```

```python
sns.histplot( x= "Volume (ml)", data = iris, kde = "True")
```

```python
sns.histplot( y= "Calories", data = iris, hue = "type")
```

```python
## Barplot : 숫자 변수의 경향 추청치를 보여준다.

sns.barplot(x="Volume (ml)", y="Calories", data =iris)

```

```python
## Box plot

sns.boxplot(x="Volume (ml)", y="Caffeine (mg)", data=iris)
```
